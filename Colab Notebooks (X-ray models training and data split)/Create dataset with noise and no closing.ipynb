```python
mainDIR = os.listdir('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)')
train_folder= '/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/train'
val_folder = '/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/val'
test_folder = '/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/test'
```

```python
# function to calculate viral/bac in a path
def stat_viral_bac(path):
  #stats train pneumonia viral/bacterial
  bac_str ="bacteria"
  cpt_bac = 0
  cpt_vir = 0
  for image_name in os.listdir(path) :
        if bac_str in image_name :
          cpt_bac += 1
        else :
          cpt_vir += 1
  return cpt_bac,cpt_vir 
```

```python
from prettytable import PrettyTable
# statestics after augmenting the dataset (closing the gap and noise)
# stats train
num_train_normal = len(os.listdir(train_folder+'/Normal'))
num_train_pneumonia = len(os.listdir('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/train/Pneumonia'))
TOT_train = num_train_normal + num_train_pneumonia
# stats test
num_test_normal = len(os.listdir('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/test/Normal'))
num_test_pneumonia = len(os.listdir('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/test/Pneumonia'))
TOT_test = num_test_pneumonia + num_test_normal
# stats valdidation
num_valid_normal = len(os.listdir('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/val/Normal'))
num_valid_pneumonia = len(os.listdir('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/val/Pneumonia'))
TOT_valid = num_valid_normal + num_valid_pneumonia

# stats train viral/bac
num_train_bac,num_train_vir = stat_viral_bac('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/train/Pneumonia')
# stats test viral/bac
num_test_bac,num_test_vir = stat_viral_bac('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/test/Pneumonia')
# stats validation viral/bac
num_val_bac,num_val_vir = stat_viral_bac('/content/drive/MyDrive/No Aug (noise) (splited .7 .15 .15)/val/Pneumonia')

TOT_train_vb = num_train_bac +num_train_vir
TOT_test_vb = num_test_bac + num_test_vir
TOT_valid_vb = num_val_bac + num_val_vir 
#table stat general
t = PrettyTable(['Class','Train','Validation', 'Test'])
t.add_row(['Normal',num_train_normal,num_valid_normal,num_test_normal])
t.add_row(['Pneumonia',num_train_pneumonia,num_valid_pneumonia,num_test_pneumonia])
t.add_row(['Total',TOT_train,TOT_valid,TOT_test])
print(t)

#table Pneumonia
print('\nPneumonia')
t = PrettyTable(['Class','Train','Validation' ,'Test'])
t.add_row(['Pneumonia viral',num_train_vir,num_val_vir,num_test_vir])
t.add_row(['Pneumonia bacterial',num_train_bac,num_val_bac,num_test_bac])
t.add_row(['Total',TOT_train_vb,TOT_valid_vb,TOT_test_vb])
print(t)
```

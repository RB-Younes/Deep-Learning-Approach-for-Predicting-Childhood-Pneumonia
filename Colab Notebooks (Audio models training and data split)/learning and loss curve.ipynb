```python
import re
import matplotlib.pyplot as plt

text = '''
Training Loss: 5.3243
Training Accuracy: 0.8311
Validation Loss: 39.8875
Validation Accuracy: 0.212

Training Loss: 5.1068
Training Accuracy: 0.8465
Validation Loss: 11.218
Validation Accuracy: 0.253

Training Loss: 4.9304
Training Accuracy: 0.8593
Validation Loss: 5.3951
Validation Accuracy: 0.5385

Training Loss: 4.9142
Training Accuracy: 0.8555
Validation Loss: 4.508
Validation Accuracy: 0.8598

Training Loss: 4.897
Training Accuracy: 0.8568
Validation Loss: 4.5272
Validation Accuracy: 0.8682
'''

pattern = r'Training Loss: (\d+\.\d+)\nTraining Accuracy: (\d+\.\d+)\nValidation Loss: (\d+\.\d+)\nValidation Accuracy: (\d+\.\d+)'
matches = re.findall(pattern, text)

if matches:
    training_losses = []
    training_accuracies = []
    validation_losses = []
    validation_accuracies = []

    for match in matches:
        training_loss, training_accuracy, validation_loss, validation_accuracy = map(float, match)
        training_losses.append(training_loss)
        training_accuracies.append(training_accuracy)
        validation_losses.append(validation_loss)
        validation_accuracies.append(validation_accuracy)

    # Plotting
    plt.figure(figsize=(16, 6))

    # Plot Training Loss and Validation Loss
    plt.subplot(1, 2, 1)
    epochs = range(1, len(matches) + 1)
    plt.plot(epochs, training_losses, marker='o', label='Training Loss')
    plt.plot(epochs, validation_losses, marker='o', label='Validation Loss')
    plt.xlabel('Folds')
    plt.ylabel('Loss')
    plt.title('Training Loss and Validation Loss per fold')
    plt.legend()

    # Plot Training Accuracy and Validation Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, training_accuracies, marker='o', label='Training Accuracy')
    plt.plot(epochs, validation_accuracies, marker='o', label='Validation Accuracy')
    plt.xlabel('Folds')
    plt.ylabel('Accuracy')
    plt.title('Training Accuracy and Validation Accuracy per fold')
    plt.legend()

    plt.tight_layout()
    plt.savefig('plot.png')
    plt.show()

else:
    print("No matches found.")

```

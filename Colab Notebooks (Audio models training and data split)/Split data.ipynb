# **Stats Folders**

```python
from prettytable import PrettyTable
```

```python
num_train_wav = len(os.listdir(train_wav))
num_train_json = len(os.listdir(train_json))
num_test_intra_json = len(os.listdir(test_intra_json))
num_test_inter_json = len(os.listdir(test_inter_json))
num_test_wav = len(os.listdir(test_wav))

num_test_json = num_test_inter_json + num_test_intra_json

tot_wav = num_train_wav + num_test_wav
tot_json = num_test_json + num_train_json
#table stat general
t = PrettyTable(['Class','Train_wav','Train_json', 'Test_intra_json','Test_inter_json','Test_json','Test_wav'])
t.add_row(['number',num_train_wav,num_train_json,num_test_intra_json,num_test_inter_json,num_test_json,num_test_wav])
print(t)

t = PrettyTable(['Class','Total wav','Total json'])
t.add_row(['number',tot_wav,tot_json])
print(t)
```

# **Stats classes**

```python
import json
list_of_paths = [train_json, test_intra_json, test_inter_json] 
count_n = 0 #normal
count_p = 0 #poor quality
count_a = 0 #anomaly
#train
for j_file in os.listdir(train_json) : 
  f = open(train_json + j_file )
  data = json.load(f)
  class_name = data.get('record_annotation')
  if class_name == 'Normal':
      count_n += 1
  elif class_name == 'Poor Quality':
      count_p += 1
  else:
      count_a += 1
Tot = count_n + count_p + count_a
t = PrettyTable(['Train','Normal','Anomaly','Poor Quality','Total'])
t.add_row(['number',count_n,count_a,count_p,Tot])
print(t)

count_n = 0 #normal
count_p = 0 #poor quality
count_a = 0 #anomaly
#test intra
for j_file in os.listdir(test_intra_json) : 
  f = open(test_intra_json + j_file )
  data = json.load(f)
  class_name = data.get('record_annotation')
  if class_name == 'Normal':
      count_n += 1
  elif class_name == 'Poor Quality':
      count_p += 1
  else:
      count_a += 1
Tot = count_n + count_p + count_a
t = PrettyTable(['Test intra','Normal','Anomaly','Poor Quality','Total'])
t.add_row(['number',count_n,count_a,count_p,Tot ])
print(t)

count_n = 0 #normal
count_p = 0 #poor quality
count_a = 0 #anomaly
#test intra
for j_file in os.listdir(test_inter_json) : 
  f = open(test_inter_json + j_file )
  data = json.load(f)
  class_name = data.get('record_annotation')
  if class_name == 'Normal':
      count_n += 1
  elif class_name == 'Poor Quality':
      count_p += 1
  else:
      count_a += 1
Tot = count_n + count_p + count_a
t = PrettyTable(['Test inter','Normal','Anomaly','Poor Quality','Total'])
t.add_row(['number',count_n,count_a,count_p,Tot])
print(t)
```

# **Split the data into folders**

```python
import shutil
```

```python
main_split_dir ='/content/drive/MyDrive/Sound Data Split/'
print(os.listdir(main_split_dir))
train_dir ='/content/drive/MyDrive/Sound Data Split/Train/'
print(os.listdir(train_dir))
test_intra_dir ='/content/drive/MyDrive/Sound Data Split/Test intra/'
print(os.listdir(test_intra_dir))
test_inter_dir ='/content/drive/MyDrive/Sound Data Split/Test inter/'
print(os.listdir(test_inter_dir))
```

```python
#train
source_folder = train_wav
destination_folder = train_dir
cpt = 0
for j_file in os.listdir(train_json) : 
  f = open(train_json + j_file )
  data = json.load(f)
  class_name = data.get('record_annotation')

  j_file = j_file.replace(".json", ".wav")
  source = source_folder + j_file
  if class_name == 'Normal':
      destination = destination_folder + 'Normal/'+ j_file
  elif class_name == 'Poor Quality':
      destination = destination_folder + 'Poor Quality/'+ j_file
  else:
      destination = destination_folder + 'Anomaly/'+ j_file

  if os.path.isfile(source):
        cpt+=1
        #shutil.copy(source, destination)
        print('copied', j_file)
print(cpt)
```

```python
#test inter
source_folder = test_wav
destination_folder = test_inter_dir
cpt = 0
for j_file in os.listdir(test_inter_json) : 
  f = open(test_inter_json + j_file )
  data = json.load(f)
  class_name = data.get('record_annotation')

  j_file = j_file.replace(".json", ".wav")
  source = source_folder + j_file
  if class_name == 'Normal':
      destination = destination_folder + 'Normal/'+ j_file
  elif class_name == 'Poor Quality':
      destination = destination_folder + 'Poor Quality/'+ j_file
  else:
      destination = destination_folder + 'Anomaly/'+ j_file

  if os.path.isfile(source):
        cpt+=1
        #shutil.copy(source, destination)
print(cpt)
```

```python
#test intra
source_folder = test_wav
destination_folder = test_intra_dir
cpt = 0
for j_file in os.listdir(test_intra_json) : 
  f = open(test_intra_json + j_file )
  data = json.load(f)
  class_name = data.get('record_annotation')

  j_file = j_file.replace(".json", ".wav")
  source = source_folder + j_file
  if class_name == 'Normal':
      destination = destination_folder + 'Normal/'+ j_file
  elif class_name == 'Poor Quality':
      destination = destination_folder + 'Poor Quality/'+ j_file
  else:
      destination = destination_folder + 'Anomaly/'+ j_file

  if os.path.isfile(source):
        cpt+=1
        #shutil.copy(source, destination)
print(cpt)
```
